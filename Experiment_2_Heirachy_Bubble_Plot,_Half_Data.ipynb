```python
# =========================
# 1. Install + Imports
# =========================
!pip install pandas plotly openpyxl

import pandas as pd
import plotly.express as px

# =========================
# 2. Load Data
# =========================
file_path = '/content/Protective and Risk factors - Tobacco .xlsx'
df = pd.read_excel(file_path)

# =========================
# 3. Clean Columns
# =========================
df.columns = [
    "ignore1", "Factor", "Influence Status", "Num Influence",
    "Category of Influence", "Num Category",
    "Sub Category", "Num Sub Category",
    "Notes", "Frequency"
]

# Remove empty rows
df = df[df["Factor"].notna()]
df = df[df["Num Influence"].notna()]

# Fill missing frequency
df["Frequency"] = df["Frequency"].fillna(1)

# =========================
# 4. FIX THE ERROR (important)
# =========================
# Fill missing subcategories so hierarchy is valid
df["Sub Category"] = df["Sub Category"].fillna("Unspecified")

# =========================
# 5. Add Labels + Colours
# =========================
df["Influence Label"] = df["Num Influence"].map({
    1: "Non-Smoking",
    2: "Smoking"
})

color_map = {
    "Smoking": "red",
    "Non-Smoking": "green"
}

# =========================
# 6. Group Data (better bubble sizes)
# =========================
df_grouped = df.groupby(
    ["Category of Influence", "Sub Category", "Factor", "Influence Label"],
    as_index=False
)["Frequency"].sum()

# =========================
# 7. Create Nested Bubble Chart
# =========================
fig = px.sunburst(
    df_grouped,
    path=[
        "Category of Influence",
        "Sub Category",
        "Factor"
    ],
    values="Frequency",
    color="Influence Label",
    color_discrete_map=color_map
)

fig.update_layout(
    title="Smoking vs Non-Smoking Influences",
)

fig.show()
```
