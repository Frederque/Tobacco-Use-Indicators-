```python
# =========================
# 1. Install + Imports
# =========================
!pip install pandas plotly openpyxl

import pandas as pd
import plotly.express as px

# =========================
# 2. Load Data
# =========================
file_path = '/content/Protective and Risk factors - Tobacco .xlsx'
df = pd.read_excel(file_path)

# =========================
# 3. Clean Columns
# =========================
df.columns = [
    "ignore1", "Factor", "Influence Status", "Num Influence",
    "Category of Influence", "Num Category",
    "Sub Category", "Num Sub Category",
    "Notes", "Frequency"
]

df = df[df["Factor"].notna()]
df = df[df["Num Influence"].notna()]

df["Frequency"] = df["Frequency"].fillna(1)

# =========================
# 4. Fix hierarchy issue
# =========================
df["Sub Category"] = df["Sub Category"].fillna("Unspecified")

# =========================
# 5. CLEAN LABELS (remove leading numbers)
# =========================
df["Category of Influence"] = df["Category of Influence"].str.replace(r"^\d+\s*", "", regex=True)
df["Sub Category"] = df["Sub Category"].str.replace(r"^\d+\s*", "", regex=True)

# =========================
# 6. Labels + Colours
# =========================
df["Influence Label"] = df["Num Influence"].map({
    1: "Non-Smoking",
    2: "Smoking"
})

color_map = {
    "Smoking": "red",
    "Non-Smoking": "green"
}

# =========================
# 7. Group Data
# =========================
df_grouped = df.groupby(
    ["Category of Influence", "Sub Category", "Factor", "Influence Label"],
    as_index=False
)["Frequency"].sum()

# =========================
# 8. Create Sunburst
# =========================
fig = px.sunburst(
    df_grouped,
    path=[
        "Category of Influence",
        "Sub Category",
        "Factor"
    ],
    values="Frequency",
    color="Influence Label",
    color_discrete_map=color_map
)

# =========================
# 9. FORCE CENTER CIRCLE TO GREY
# =========================
fig.update_traces(
    root_color="lightgrey"
)

# =========================
# 10. Layout
# =========================
fig.update_layout(
    title="Smoking vs Non-Smoking Influences"
)

fig.show()
```

```python
# =========================
# 1. Install + Imports
# =========================
!pip install pandas plotly openpyxl

import pandas as pd
import plotly.express as px

# =========================
# 2. Load Data
# =========================
file_path = '/content/Protective and Risk factors - Tobacco .xlsx'
df = pd.read_excel(file_path)

# =========================
# 3. Clean Columns
# =========================
df.columns = [
    "ignore1", "Factor", "Influence Status", "Num Influence",
    "Category of Influence", "Num Category",
    "Sub Category", "Num Sub Category",
    "Notes", "Frequency"
]

df = df[df["Factor"].notna()]
df = df[df["Num Influence"].notna()]
df["Frequency"] = df["Frequency"].fillna(1)

# =========================
# 4. Fix hierarchy issue
# =========================
df["Sub Category"] = df["Sub Category"].fillna("Unspecified")

# =========================
# 5. CLEAN LABELS
# =========================
df["Category of Influence"] = df["Category of Influence"].str.replace(r"^\d+\s*", "", regex=True)
df["Sub Category"] = df["Sub Category"].str.replace(r"^\d+\s*", "", regex=True)

# =========================
# 6. Labels + Colours
# =========================
df["Influence Label"] = df["Num Influence"].map({
    1: "Non-Smoking",
    2: "Smoking"
})

color_map = {
    "Smoking": "red",
    "Non-Smoking": "green"
}

# =========================
# 7. Group Data
# =========================
df_grouped = df.groupby(
    ["Influence Label", "Category of Influence", "Sub Category", "Factor"],
    as_index=False
)["Frequency"].sum()

# =========================
# 8. Add root
# =========================
df_grouped["All"] = "All Influences"

# =========================
# 9. Create Sunburst
# =========================
fig = px.sunburst(
    df_grouped,
    path=[
        "All",
        "Influence Label",
        "Category of Influence",
        "Sub Category",
        "Factor"
    ],
    values="Frequency",
    color="Influence Label",
    color_discrete_map=color_map
)

# =========================
# 10. Fix TEXT + SIZE
# =========================
fig.update_traces(
    root_color="lightgrey",
    textinfo="label+percent entry",
    insidetextorientation="radial"  # better fit
)

# =========================
# 11. Make figure BIGGER
# =========================
fig.update_layout(
    title="Smoking vs Non-Smoking Influences",
    width=1000,     # increase size
    height=1000,
    margin=dict(t=80, l=20, r=20, b=20)
)

fig.show()
```
